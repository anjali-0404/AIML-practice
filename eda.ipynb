## 1. Load and Inspect Dataset

Import required libraries and load the Titanic dataset to understand its basic structure.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configure visualization settings
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✓ All libraries imported successfully!")
```

```python
# Load the Titanic dataset
df = pd.read_csv('../Task 1/Titanic-Dataset.csv')

# Display first 5 rows
print("📊 FIRST 5 ROWS OF THE DATASET:\n")
print(df.head())
print("\n" + "="*80)
```

```python
# Display Dataset Information
print("📋 DATASET INFORMATION:\n")
df.info()
print("\n" + "="*80)

# Display shape
print(f"\n📏 DATASET SHAPE: {df.shape}")
print(f"   • Total Records (Rows): {df.shape[0]}")
print(f"   • Total Features (Columns): {df.shape[1]}")
```

## 2. Data Understanding and Analysis

Analyze dataset dimensions, column names, data types, missing values, and unique values.

```python
# Column Names and Data Types
print("📌 COLUMN NAMES AND DATA TYPES:\n")
for col in df.columns:
    print(f"   {col}: {df[col].dtype}")

print("\n" + "="*80)

# Data Types Summary
print("\n🔍 DATA TYPES SUMMARY:")
print(df.dtypes.value_counts())

print("\n" + "="*80)
```

```python
# Missing Values Analysis
print("⚠️  MISSING VALUES ANALYSIS:\n")
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing,
    'Percentage': missing_percent
})
print(missing_df[missing_df['Missing_Count'] > 0].sort_values('Percentage', ascending=False))

print("\n" + "="*80)

# Unique Values
print("\n🎯 UNIQUE VALUES PER COLUMN:\n")
unique_info = df.nunique()
print(unique_info)
```

## 3. Summary Statistics

Generate comprehensive statistical summaries for numerical features.

```python
# Descriptive Statistics
print("📊 DESCRIPTIVE STATISTICS:\n")
print(df.describe().round(3))

print("\n" + "="*80)
```

```python
# Mean, Median, and Standard Deviation
print("📈 CENTRAL TENDENCY AND DISPERSION MEASURES:\n")

stats_table = pd.DataFrame({
    'Mean': df.mean(numeric_only=True).round(3),
    'Median': df.median(numeric_only=True).round(3),
    'Std Dev': df.std(numeric_only=True).round(3),
    'Min': df.min(numeric_only=True).round(3),
    'Max': df.max(numeric_only=True).round(3)
})

print(stats_table)

print("\n" + "="*80)

# Quartiles Analysis
print("\n📊 QUARTILES ANALYSIS (25%, 50%, 75%):\n")
print(df.quantile([0.25, 0.5, 0.75]).round(3))
```

## 4. Data Cleaning and Preprocessing

Handle missing values, remove duplicates, and prepare data for analysis.

```python
# Create a copy for cleaning operations
df_clean = df.copy()

print("🔧 DATA CLEANING PROCESS:\n")

# Check for duplicates
print(f"1. Duplicate rows before removal: {df_clean.duplicated().sum()}")
df_clean = df_clean.drop_duplicates()
print(f"   ✓ Duplicate rows after removal: {df_clean.duplicated().sum()}")

# Handle missing values
print(f"\n2. Handling Missing Values:")
print(f"   • Age: {df_clean['Age'].isnull().sum()} missing values")
print(f"   • Cabin: {df_clean['Cabin'].isnull().sum()} missing values")
print(f"   • Embarked: {df_clean['Embarked'].isnull().sum()} missing values")

# Fill Age with median (preserves distribution)
age_median = df_clean['Age'].median()
df_clean['Age'].fillna(age_median, inplace=True)
print(f"   ✓ Age filled with median ({age_median:.1f})")

# Fill Embarked with mode (most frequent port)
embarked_mode = df_clean['Embarked'].mode()[0]
df_clean['Embarked'].fillna(embarked_mode, inplace=True)
print(f"   ✓ Embarked filled with mode ({embarked_mode})")

# Drop Cabin (too many missing values - 77%)
df_clean.drop('Cabin', axis=1, inplace=True)
print(f"   ✓ Cabin column dropped (77% missing)")

print(f"\n3. Final Dataset Status:")
print(f"   • Shape after cleaning: {df_clean.shape}")
print(f"   • Total missing values: {df_clean.isnull().sum().sum()}")
print(f"   • Data quality: 100% complete ✓")
```

## 5. Univariate Analysis with Histograms

Analyze distribution patterns of Age and Fare variables.

```python
# Create histograms for Age and Fare distributions
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Distribution Analysis: Age and Fare', fontsize=16, fontweight='bold', y=1.02)

# Age Distribution Histogram
axes[0].hist(df_clean['Age'], bins=30, color='steelblue', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Age (years)', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Frequency', fontsize=11, fontweight='bold')
axes[0].set_title('Age Distribution', fontsize=12, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)

# Add statistical lines
age_mean = df_clean['Age'].mean()
age_median = df_clean['Age'].median()
axes[0].axvline(age_mean, color='red', linestyle='--', linewidth=2, label=f'Mean: {age_mean:.1f}')
axes[0].axvline(age_median, color='green', linestyle='--', linewidth=2, label=f'Median: {age_median:.1f}')
axes[0].legend(fontsize=10)

# Fare Distribution Histogram
axes[1].hist(df_clean['Fare'], bins=30, color='coral', edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Fare ($)', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Frequency', fontsize=11, fontweight='bold')
axes[1].set_title('Fare Distribution', fontsize=12, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

# Add statistical lines
fare_mean = df_clean['Fare'].mean()
fare_median = df_clean['Fare'].median()
axes[1].axvline(fare_mean, color='red', linestyle='--', linewidth=2, label=f'Mean: {fare_mean:.2f}')
axes[1].axvline(fare_median, color='green', linestyle='--', linewidth=2, label=f'Median: {fare_median:.2f}')
axes[1].legend(fontsize=10)

plt.tight_layout()
plt.savefig('histograms.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Histograms displayed and saved")
```

### Key Observations:
- **Age Distribution**: Approximately normal distribution with slight right skew. Most passengers were between 20-40 years old. Children (0-5) are visible in the data.
- **Fare Distribution**: Highly right-skewed with most fares concentrated below $100, but outliers exist for premium fares (up to $500+).

## 6. Distribution Analysis with Boxplots

Visualize relationships between Age/Fare and categorical variables using boxplots.

```python
# Create boxplots for Age and Fare analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Boxplot Analysis: Distribution Across Categories', fontsize=16, fontweight='bold', y=1.02)

# Age vs Survival boxplot
df_clean.boxplot(column='Age', by='Survived', ax=axes[0], patch_artist=True)
axes[0].set_xlabel('Survival Status', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Age (years)', fontsize=11, fontweight='bold')
axes[0].set_title('Age Distribution by Survival Status', fontsize=12, fontweight='bold')
axes[0].set_xticklabels(['Did Not Survive', 'Survived'])
plt.sca(axes[0])
axes[0].get_figure().suptitle('')  # Remove automatic title

# Fare vs Passenger Class boxplot
df_clean.boxplot(column='Fare', by='Pclass', ax=axes[1], patch_artist=True)
axes[1].set_xlabel('Passenger Class', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Fare ($)', fontsize=11, fontweight='bold')
axes[1].set_title('Fare Distribution by Passenger Class', fontsize=12, fontweight='bold')
axes[1].set_xticklabels(['1st Class', '2nd Class', '3rd Class'])
plt.sca(axes[1])
axes[1].get_figure().suptitle('')  # Remove automatic title

plt.tight_layout()
plt.savefig('boxplots.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Boxplots displayed and saved")
```

### Key Observations:
- **Age vs Survival**: Survivors had a slightly lower median age, suggesting children and younger people had better survival chances
- **Fare vs Class**: First class passengers paid significantly higher fares, with median around $50-60 compared to 3rd class (~$8-10)

## 7. Categorical Analysis with Countplots

Analyze categorical variables and their relationships with survival.

```python
# Create countplots for categorical analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Categorical Distribution Analysis', fontsize=16, fontweight='bold', y=1.02)

# Gender vs Survival
sex_survived = pd.crosstab(df_clean['Sex'], df_clean['Survived'])
sex_survived.plot(kind='bar', ax=axes[0], color=['#d62728', '#2ca02c'], edgecolor='black', alpha=0.8)
axes[0].set_xlabel('Gender', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Count', fontsize=11, fontweight='bold')
axes[0].set_title('Gender Distribution by Survival Status', fontsize=12, fontweight='bold')
axes[0].legend(['Did Not Survive', 'Survived'], loc='upper right', fontsize=10)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)
axes[0].grid(axis='y', alpha=0.3)

# Passenger Class Distribution
pclass_counts = df_clean['Pclass'].value_counts().sort_index()
axes[1].bar(pclass_counts.index, pclass_counts.values, 
            color=['#1f77b4', '#ff7f0e', '#2ca02c'], edgecolor='black', alpha=0.8)
axes[1].set_xlabel('Passenger Class', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Count', fontsize=11, fontweight='bold')
axes[1].set_title('Passenger Class Distribution', fontsize=12, fontweight='bold')
axes[1].set_xticks([1, 2, 3])
axes[1].set_xticklabels(['1st Class', '2nd Class', '3rd Class'])
axes[1].grid(axis='y', alpha=0.3)

# Add counts on bars
for i, v in enumerate(pclass_counts.values):
    axes[1].text(pclass_counts.index[i], v + 5, str(v), ha='center', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.savefig('countplots.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Countplots displayed and saved")
```

```python
# Calculate survival rates
print("📊 SURVIVAL RATE ANALYSIS:\n")

female_survival = (df_clean[df_clean['Sex'] == 'female']['Survived'].sum() / 
                   len(df_clean[df_clean['Sex'] == 'female'])) * 100
male_survival = (df_clean[df_clean['Sex'] == 'male']['Survived'].sum() / 
                 len(df_clean[df_clean['Sex'] == 'male'])) * 100

print(f"Female Survival Rate: {female_survival:.2f}%")
print(f"Male Survival Rate: {male_survival:.2f}%")
print(f"\nPassenger Class Survival Rates:")

for pclass in [1, 2, 3]:
    surv_rate = (df_clean[df_clean['Pclass'] == pclass]['Survived'].sum() / 
                 len(df_clean[df_clean['Pclass'] == pclass])) * 100
    count = len(df_clean[df_clean['Pclass'] == pclass])
    print(f"  • Class {pclass}: {surv_rate:.2f}% ({count} passengers)")
```

### Key Observations:
- **Gender Impact**: Female passengers had ~74% survival rate vs. ~19% for males - "Women and Children First" policy was clearly followed
- **Class Impact**: 1st class had ~63% survival, 2nd class ~47%, 3rd class ~24% - socioeconomic status was crucial

## 8. Correlation Analysis with Heatmap

Calculate and visualize correlations between numerical features.

```python
# Calculate correlation matrix
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
correlation_matrix = df_clean[numeric_cols].corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=1.5, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1, annot_kws={'fontsize': 11, 'fontweight': 'bold'})
plt.title('Correlation Matrix - Titanic Dataset', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Correlation heatmap displayed and saved")
```

```python
# Analyze key correlations with Survival
print("🔗 KEY CORRELATIONS WITH SURVIVAL:\n")
survival_corr = correlation_matrix['Survived'].sort_values(ascending=False)
print(survival_corr)

print("\n📝 INTERPRETATION:")
print(f"   • Pclass (-0.34): Negative correlation - Higher class number (lower status) → Lower survival")
print(f"   • Fare (0.26): Positive correlation - Higher fare → Higher survival")
print(f"   • Age (-0.07): Weak negative - Younger passengers had slightly higher survival")
```

## 10. Pattern Detection and Outlier Analysis

Identify trends, outliers, and important patterns in the data.

```python
# Overall Survival Statistics
print("📈 SURVIVAL OVERVIEW:\n")
survival_rate = (df_clean['Survived'].sum() / len(df_clean)) * 100
print(f"   • Overall Survival Rate: {survival_rate:.2f}%")
print(f"   • Survived: {df_clean['Survived'].sum()} passengers")
print(f"   • Did Not Survive: {(df_clean['Survived'] == 0).sum()} passengers")

# Outlier Detection using IQR method
print("\n🔍 OUTLIER DETECTION (IQR Method):\n")

# Age outliers
Q1_age = df_clean['Age'].quantile(0.25)
Q3_age = df_clean['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age
outliers_age = df_clean[(df_clean['Age'] < lower_bound_age) | (df_clean['Age'] > upper_bound_age)]

print(f"Age Outliers:")
print(f"   • Lower Bound: {lower_bound_age:.2f}, Upper Bound: {upper_bound_age:.2f}")
print(f"   • Number of Outliers: {len(outliers_age)} ({(len(outliers_age)/len(df_clean)*100):.2f}%)")

# Fare outliers
Q1_fare = df_clean['Fare'].quantile(0.25)
Q3_fare = df_clean['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
lower_bound_fare = Q1_fare - 1.5 * IQR_fare
upper_bound_fare = Q3_fare + 1.5 * IQR_fare
outliers_fare = df_clean[(df_clean['Fare'] < lower_bound_fare) | (df_clean['Fare'] > upper_bound_fare)]

print(f"\nFare Outliers:")
print(f"   • Lower Bound: {lower_bound_fare:.2f}, Upper Bound: {upper_bound_fare:.2f}")
print(f"   • Number of Outliers: {len(outliers_fare)} ({(len(outliers_fare)/len(df_clean)*100):.2f}%)")
print(f"   • Min Outlier Fare: ${outliers_fare['Fare'].min():.2f}")
print(f"   • Max Outlier Fare: ${outliers_fare['Fare'].max():.2f}")
```

```python
# Age group analysis
print("\n\n👶 AGE GROUP ANALYSIS:\n")
age_groups = pd.cut(df_clean['Age'], bins=[0, 5, 12, 18, 35, 60, 100], 
                     labels=['0-5', '6-12', '13-18', '19-35', '36-60', '60+'])
age_group_survival = pd.crosstab(age_groups, df_clean['Survived'], margins=True)
print(age_group_survival)

# Feature Importance based on correlation
print("\n\n⭐ FEATURE IMPORTANCE (Correlation with Survival):\n")
importance = correlation_matrix['Survived'].drop('Survived').abs().sort_values(ascending=False)
for i, (feature, corr) in enumerate(importance.items(), 1):
    print(f"   {i}. {feature}: {corr:.4f}")
    
# Embarked port analysis
print("\n\n🚢 EMBARKED PORT ANALYSIS:\n")
embarked_counts = df_clean['Embarked'].value_counts()
port_names = {'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'}
for port, count in embarked_counts.items():
    percent = (count / len(df_clean)) * 100
    print(f"   • {port} ({port_names.get(port, port)}): {count} passengers ({percent:.1f}%)")
```

## 11. Key Observations and Insights

### Major Findings:

**1. Primary Survival Predictors (in order of importance):**
   - **Gender**: Female passengers had 74% survival vs 19% for males
   - **Passenger Class**: 1st class 63%, 2nd class 47%, 3rd class 24%
   - **Age**: Younger passengers, especially children, had better survival chances
   - **Fare**: Higher paying passengers had better survival rates

**2. Critical Patterns Identified:**
   - ✓ "Women and Children First" evacuation policy was strictly followed
   - ✓ Strong socioeconomic bias in survival (class/fare correlation)
   - ✓ Age distribution shows working-age population (20-40) was largest passenger group
   - ✓ Fare distribution is right-skewed with numerous luxury passengers (outliers)

**3. Data Quality Insights:**
   - Quality dataset with 891 records and clean structure
   - Missing values (Age, Cabin, Embarked) handled appropriately
   - No duplicates present
   - Clear categorical and numerical feature separation

**4. Outliers and Special Cases:**
   - ~27% of passengers had fares above Q3 (premium travelers)
   - Large age range (0-80 years) providing diverse demographic
   - Cabin feature dropped due to 77% missing values

### Recommendations for Predictive Modeling:

1. **Feature Engineering Suggestions:**
   - Create family size feature (SibSp + Parch)
   - Extract title from passenger names (Mr, Mrs, Miss, Dr, etc.)
   - Group ages into meaningful bins (Child, Teen, Adult, Senior)
   - Encode categorical variables (Sex, Embarked)

2. **Model Considerations:**
   - Account for class imbalance (62% did not survive)
   - Use feature scaling for Age and Fare
   - Consider feature importance ranking: Pclass > Fare > Age > Sex
   - Cross-validation essential for robust performance estimation

3. **Potential ML Algorithms:**
   - Classification models: Logistic Regression, Random Forest, Gradient Boosting
   - Ensemble methods recommended for better accuracy
   - Evaluate using ROC-AUC, Precision-Recall curves
   - Address imbalance with SMOTE or class weights

The Titanic dataset reveals a tragic human story with clear patterns of discrimination and social hierarchy. The analysis demonstrates that survival was primarily determined by gender and socioeconomic status rather than random chance. These insights provide a foundation for building predictive models and understanding how social factors influenced life-and-death situations during historical disasters.
