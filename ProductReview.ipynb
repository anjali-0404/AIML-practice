```python
# =========================================
# 🧠 PRODUCT REVIEW SUMMARY GENERATOR
# Using Pretrained T5 Transformer Model
# =========================================

# 📦 Import Required Libraries
!pip install transformers datasets sentencepiece --quiet

from transformers import T5Tokenizer, T5ForConditionalGeneration
from datasets import load_dataset
import torch

# ===============================
# 📥 Load Predefined Dataset
# ===============================
# We'll use a smaller predefined dataset for Amazon Fine Food Reviews
# (available in the HuggingFace 'amazon_polarity' dataset)
dataset = load_dataset("amazon_polarity", split="train[:2000]")  # small subset for demo

# Display sample data
print("Sample Review:\n", dataset[0]['content'])
print("\nSample Label (1=Positive, 0=Negative):", dataset[0]['label'])

# ===============================
# 🧹 Preprocess Reviews
# ===============================
# We'll combine reviews into text input for summarization
reviews = [item['content'] for item in dataset.select(range(5))]  # use first 5 samples

# ===============================
# 🤖 Load Pretrained T5 Model
# ===============================
model_name = "t5-small"  # lightweight version of T5
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# ===============================
# 📝 Summarize Each Review
# ===============================
for i, review in enumerate(reviews):
    # Prepare input for T5
    input_text = "summarize: " + review
    inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)

    # Generate summary
    summary_ids = model.generate(
        inputs,
        max_length=60,  # length of summary
        min_length=10,
        length_penalty=2.0,
        num_beams=4,
        early_stopping=True
    )

    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    print(f"\n🔹 Review {i+1}:\n{review[:400]}...")
    print(f"✨ Summary {i+1}: {summary}")
    print("="*80)

# ===============================
# 🧾 Custom Review Test
# ===============================
custom_review = """
I bought this product last month and I am very satisfied with its performance.
The sound quality is crystal clear, and battery life is outstanding.
However, the price could have been a bit lower. Overall, I would highly recommend it.
"""

input_text = "summarize: " + custom_review
inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)

summary_ids = model.generate(inputs, max_length=60, min_length=10, num_beams=4, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("\n🗣️ Custom Review Summary:\n", summary)
```
