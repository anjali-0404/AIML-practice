```python
import numpy as np
```

```python
# Step 1: Define the step activation function
def step_function(x):
    return 1 if x >= 0 else 0
```

```python
# Step 2: Perceptron training function
def train_perceptron(X, y, lr=0.1, epochs=10):
    # Initialize weights and bias
    weights = np.zeros(X.shape[1])
    bias = 0

    for _ in range(epochs):
        for x, target in zip(X, y):
            linear_output = np.dot(weights, x) + bias
            prediction = step_function(linear_output)
            error = target - prediction
            weights += lr * error * x
            bias += lr * error
    return weights, bias
```

```python
# Step 3: Prediction function
def predict(X, weights, bias):
    return [step_function(np.dot(weights, xi) + bias) for xi in X]
```

```python
# Step 4: Example usage with AND gate
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

weights, bias = train_perceptron(X, y)
```

```python
# Test predictions
for xi in X:
    print(f"Input: {xi}, Predicted: {predict([xi], weights, bias)[0]}")
```
